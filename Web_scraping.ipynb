#Introduction
I asked myself in my country what is the most used web apps and from a short survey i conducted, (amongst friends and family) i found out that it was e-commerce web apps.
Since now they constitute to helping make life easier i decided to take a look and develop more precise data for data scientist to analyse

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.brandnav.io/reports/countries/ghana/'

page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all('th')
ecom = soup.find_all('th')
```

```python
ecom
```

```python
ecom_title = [title.text for title in ecom]
```

```python
print(ecom_title)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = ecom_title)

df
```

```python
column_data = soup.find_all('tr')
```

```python
for row in column_data [1:]:
    row_data = row.find_all('td')
    individual_data = [data.text for data in row_data]


    length = len(df)
    df.loc[length] = individual_data
```

```python
df
```

```python
df = df.drop(['Rank','Support Emails','City','Support Phone Number'], axis = 1)
```

```python
df
```

```python
import os
output_dir = '/content/downloads/'
os.makedirs(output_dir, exist_ok=True)
df.to_csv(os.path.join(output_dir, 'E-com.csv'), index=False)
```
